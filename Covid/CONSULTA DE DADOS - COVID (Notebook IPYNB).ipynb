```
%sql

DESCRIBE
  default.tb_paciente
```

```
%sql

SELECT
  sexo, 
CASE
  WHEN febre = 1 THEN 'sim'
  WHEN febre = 2 THEN 'nao'
  ELSE 'ignorado'
END AS teve_febre, count(febre) AS count
FROM 
  tb_paciente
WHERE 
  febre IS NOT null
GROUP BY
  sexo, teve_febre
ORDER BY
  sexo, teve_febre asc;
```

```
%sql

SELECT
  sexo, 
CASE
  WHEN tosse = 1 THEN 'sim'
  ELSE 'nao'
END AS teve_tosse, count(tosse) AS count
FROM 
  tb_paciente
WHERE 
  tosse IS NOT null
GROUP BY
  sexo, teve_tosse
ORDER BY
  sexo, teve_tosse asc;
```

```
%sql

SELECT
  sexo, 
CASE
  WHEN hospital = 1 THEN 'sim'
  ELSE 'nao'
END AS paciente_internado, count(hospital) AS count
FROM 
  tb_paciente
WHERE 
  hospital IS NOT null
GROUP BY
  sexo, paciente_internado
ORDER BY
  sexo, paciente_internado asc;
```

```
%sql

SELECT 
  count(id) AS qt_pacientes_internados
FROM
  default.tb_paciente
WHERE
  sexo IS NOT NULL AND dt_interna IS NOT NULL AND dt_evoluca IS NOT NULL;
```

```
%sql

SELECT
  sexo, dt_interna, dt_evoluca, date_diff(dt_evoluca, dt_interna) AS dias_evolucao
FROM 
  tb_paciente
WHERE
  dt_interna IS NOT NULL AND dt_evoluca IS NOT NULL
GROUP BY
  sexo, dt_interna, dt_evoluca, dias_evolucao
ORDER BY
  sexo asc;
```

```
%sql

SELECT
    sexo, MAX(date_diff(dt_evoluca, dt_interna)) as dias_evolucao_maior
FROM
  tb_paciente
WHERE
  dt_interna IS NOT NULL AND dt_evoluca IS NOT NULL
GROUP BY
  sexo;
```

```
%sql

SELECT
    sexo, MIN(date_diff(dt_evoluca, dt_interna)) as dias_evolucao_maior
FROM
  tb_paciente
WHERE
  dt_interna IS NOT NULL AND dt_evoluca IS NOT NULL AND date_diff(dt_evoluca, dt_interna) > 0
GROUP BY
  sexo;
```

```
%sql

SELECT 
CASE
  WHEN cs_raca = 1 THEN 'branca'
  WHEN cs_raca = 2 THEN 'preta'
  WHEN cs_raca = 3 THEN 'amarela'
  WHEN cs_raca = 4 THEN 'parda'
  WHEN cs_raca = 5 THEN 'indigena'
  ELSE 'ignorado'
  END AS raca, sexo, count(cs_raca) AS count
FROM
  default.tb_paciente
WHERE
  sexo IS NOT NULL AND cs_raca IS NOT NULL
GROUP BY
  raca, sexo
ORDER BY
  raca asc;
```
